```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
FILE = './METR-LA/metr-la.h5'

with pd.HDFStore(FILE) as store:
    print(store.keys())
    LA = store['df']

LA.replace(0, np.nan, inplace=True)
```

```python
location = 100
sensor = LA.columns[location]
line = LA[sensor]

fig = plt.figure(figsize=(6, 4))
plt.ylim([0, 75])
plt.plot(LA.index[3000:4000], line[3000:4000], linewidth=1)
plt.ylabel('Speed (mph)')
plt.xticks(rotation=30, ha='right')
plt.title(f"Time series of traffic speed at location {sensor}")
```

```python
fig = plt.figure(figsize=(6, 4))
plt.hist(line, bins = 75)
plt.ylabel('count')
plt.xlabel('Speed (mph)')
plt.xlim([0, 75])

plt.title(f"Histogram of traffic speed at location {sensor}")
```
